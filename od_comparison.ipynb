```python
import mir_eval
import os
import json
import numpy as np
from scipy.io.wavfile import read
import time
from util import getCloseOnsets, getOnsets, summarizeScores, recalcScore, separateFiles
import quickPlayer
```

# GuitarSet

```python
soundDir = "guitarset/"
annotationDir = soundDir + "Annotations/"
solofiles,chordfiles = separateFiles(soundDir)

with open(soundDir+"list_files.txt","r") as f:
        filelist = json.load(f)
```

# CNN Onset Detector

```python
from madmom.features.onsets import OnsetPeakPickingProcessor,CNNOnsetProcessor
```

```python
scores = []
for i in range(360):
    filename = filelist[i]
    filedir = annotationDir + filename
   
    fs, x = read(soundDir + filename+ ".wav")
    gt, gtgroups = getOnsets(filedir,fs=fs)
    pred = OnsetPeakPickingProcessor()(CNNOnsetProcessor()(x))
    
    score = mir_eval.onset.evaluate(gt,pred,window=0.05)
    
    print(filename)
    print(score)
    
    scores.append([filename,score,gt,pred])
    
np.save("scores-CNN-GS",scores)
#scores = np.load("scores-CNN-GS.npy")
```

```python
summarizeScores(recalcScore(scores),chordfiles)
```

# SF-HA 

```python
import processing
```

```python
scores = []

s = time.time()
for i in range(360):
    filename = os.listdir(annotationDir)[i]
    filedir = annotationDir + filename
    
    fs, x = read(soundDir + filename+ ".wav")
    gt,gtgroups = getOnsets(filedir,dataset="gs",fs=fs,combine_window=0.05)
    
    result,candid,freqs = processing.od(soundDir + filename+ ".wav",hc=-3,div=10,L=20,a=51,res="low")
    
    score = mir_eval.onset.evaluate(gt,result+0.03,window = 0.05)
    sf_score = mir_eval.onset.evaluate(gt,candid+0.03,window = 0.05)
    print(filename)
    print(sf_score)
    print(score)
    scores.append([filename,score,gt,result,sf_score,candid,freqs])
    print(time.time()-s)
    if i%10 == 0:
        print("Saving... File number: %d"%i)
        print("Time elapsed: %d"%(time.time()-s))
        np.save("scores-SFHA-GS_low",scores)
        print("summary")
        summarizeScores(recalcScore(scores,offset=0.03),chordfiles)
        print("\n")
# scores = np.load("scores-SFHA-GS_low.npy",allow_pickle=True)   
```

```python
summarizeScores(recalcScore(scores,offset=0.03),chordfiles)
```

# SFHA

```python
import processing
```

```python
s = time.time()
scores = []
for i in range(232):
    filename = filelist[i]
    filedir = annotationDir + filename
    
    
    fs, x = read(soundDir + filename+ ".wav")
    gt,gtgroups = getOnsets(filedir,dataset="mc",fs=fs,combine_window=0.025)
    
    result,candids,freqs = processing.od(soundDir + filename+ ".wav",hc=-2,div=15,L=15,a=51,res="med")
    
    score = mir_eval.onset.evaluate(gt,result+0.01,window = 0.05)
    sf_score = mir_eval.onset.evaluate(gt,candids+0.01,window = 0.05)
     
    
    print(filename)
    print(sf_score)
    print(score)
    
    scores.append([filename,score,gt,result,sf_score,candids,freqs])
    if i%10 ==0:
        np.save("scores-SFHA-MC",scores)
        summarizeScores(recalcScore(scores,offset=0.01),chordfiles)
        print(time.time()-s)

np.save("scores-SFHA-MC",scores)
```
